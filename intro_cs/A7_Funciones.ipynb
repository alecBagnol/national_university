1. <h4>3-12 Cotejo de n-gramas</h4> <i>(Pag 109) </i>

```python
# esta funcion toma el valor n y las cadenas de texto; las cadenas las asigna a una lista
# por cada cadena recorre para obtener los ngramas teniendo el paso el número n
# centrada en el rango a determinar de la cadena de texto
def ngrama(e, a, b):
  data = [a,b]
  data_ngrama = []
  for text in data:
    splitted = []
    for index in range(0,len(text)-1):
      splitted.append(text[index:index+e])
    data_ngrama.append(splitted)
  return data_ngrama

# simplemente divide cada carácter dentro de la cadena y devuelve la lista
def to_letters(a):
  b = list(a)
  return b

# toma los casos de ngramas o caracteres y dado uno de los dos cuenta cuantas veces
# una letra de una cadena está en la otra, pero toma como condicional que no
# se repita para crear duplicados de coincidencias en la variable que va guardando
# los comunes
def matching(e):
  a = e[0]
  b = e[1]
  matches = []
  for case in b:
    count_a = a.count(case)
    if count_a > 0 and matches.count(case) < count_a:
      matches.append(case)
  return matches

# para calcular el coeficiente de Dice dada la formula en el problema
def coeficiente_dice(comun,elemento_a,elemento_b):
  coef = (2*comun) / (elemento_a + elemento_b)
  return coef

# para evitar que la persona ingrese un número n menor al tamaño de las cadenas
def entrada_n():
  # error = True
  while True: 
    n = int(input("Ahora ingrese el número n para hallar n-gramas: "))
    if n > size_a or n > size_b:
      print("Por favor, ingrese un número menor para n: ")
    else:
      return n


texto_a = input("Ingrese una palabra: ")
texto_b = input("Ingrese otra palabra: ")
size_a = len(texto_a)
size_b = len(texto_b)
n = entrada_n()
segmented = ngrama(n,texto_a, texto_b)
letters = [to_letters(texto_a),to_letters(texto_b)]
letters_match = len(matching(letters))
ngrama_match = len(matching(segmented))
coeficiente_char = coeficiente_dice(letters_match, size_a,size_b)
coeficiente_ngrama = coeficiente_dice(ngrama_match, len(segmented[0]),len(segmented[1]))
print("")
print(f"El número de n-gramas en común son: {ngrama_match}")
print(f"El coeficiente de Dice de n-gramas es : {coeficiente_ngrama}")
print(f"El número de caracteres en común son: {letters_match}")
print(f"El coeficiente de Dice de caracteres es : {coeficiente_char}")
```
